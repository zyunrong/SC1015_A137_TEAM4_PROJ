# Import Necessary Libraries and Modules

```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import mean_squared_error
from sklearn.tree import plot_tree

data = pd.read_csv('breast-cancer.csv')
df = pd.DataFrame(data)
newdata = data[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]
```

# Histogram of Benign vs Malignant Tumours


```python
px.histogram(data_frame = newdata, x = "diagnosis", color = 'diagnosis', color_discrete_sequence= ['#0000FF', '#FFA500'])
```

# Pair Plot between every other variable in the cleaned dataset

```python
# Create a pair plot with a color-coded diagnosis
sn.pairplot(newdata, hue='diagnosis', markers=["o", "s"])
plt.show()
```

# We chose the most prominent plots, i.e those with a high correlation between the variables, and showed this through a scatterplot

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

sn.scatterplot(x='radius_mean', y='perimeter_mean', hue='diagnosis', data=newdata, markers=["o", "s"], ax=axes[0, 0])
sn.scatterplot(x='radius_mean', y='area_mean', hue='diagnosis', data=newdata, markers=["o", "s"], ax=axes[0, 1])
sn.scatterplot(x='perimeter_mean', y='area_mean', hue='diagnosis', data=newdata, markers=["o", "s"], ax=axes[1, 0])
sn.scatterplot(x='compactness_mean', y='concavity_mean', hue='diagnosis', data=newdata, markers=["o", "s"], ax=axes[1, 1])
sn.scatterplot(x='compactness_mean', y='concave points_mean', hue='diagnosis', data=newdata, markers=["o", "s"], ax=axes[2, 0])

# Remove the unused subplot
axes[2, 1].axis('off')
plt.show()
```

### Boxplots and Violinplots: To further investigate the distribution of values within each feature and their correlation with the diagnosis, we generated box plots and violin plots for the features "radius_mean," "perimeter_mean," "area_mean," "compactness_mean," and "concave points_mean." The violin plots showed the expected probability density of the data at various levels, while the box plots disclosed the median, quartiles, and existence of outliers for each characteristic. 


```python
#Generate a box plot for each feature
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

sn.boxplot(x='diagnosis', y='radius_mean', data=newdata, ax=axes[0, 0])
##plt.title('Box Plot: Radius Mean')

sn.boxplot(x='diagnosis', y='perimeter_mean', data=newdata, ax=axes[0, 1])
##plt.title('Box Plot: Perimeter Mean')

sn.boxplot(x='diagnosis', y='area_mean', data=newdata, ax=axes[1, 0])
##plt.title('Box Plot: Area Mean')

sn.boxplot(x='diagnosis', y='compactness_mean', data=newdata, ax=axes[1, 1])
##plt.title('Box Plot: Compactness Mean')

sn.boxplot(x='diagnosis', y='concavity_mean', data=newdata, ax=axes[2, 0])
##plt.title('Box Plot: Concavity Mean')

sn.boxplot(x='diagnosis', y='concave points_mean', data=newdata, ax=axes[2, 1])
##plt.title('Box Plot: Concave Points Mean')

plt.show()
```

```python
# Generate violin plots for each feature
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
sn.violinplot(x='diagnosis', y='radius_mean', data=newdata, ax=axes[0, 0])
##plt.title('Violin Plot: Radius Mean')

sn.violinplot(x='diagnosis', y='perimeter_mean', data=newdata, ax=axes[0, 1])
##plt.title('Violin Plot: Perimeter Mean')

sn.violinplot(x='diagnosis', y='area_mean', data=newdata, ax=axes[1, 0])
##plt.title('Violin Plot: Area Mean')

sn.violinplot(x='diagnosis', y='compactness_mean', data=newdata, ax=axes[1, 1])
##plt.title('Violin Plot: Compactness Mean')

sn.violinplot(x='diagnosis', y='concavity_mean', data=newdata, ax=axes[2, 0])
##plt.title('Violin Plot: Concavity Mean')

sn.violinplot(x='diagnosis', y='concave points_mean', data=newdata, ax=axes[2, 1])
##plt.title('Violin Plot: Concave Points Mean')

plt.show()
```
